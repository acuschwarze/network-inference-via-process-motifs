```python
from scipy.special import hyp2f1, binom

def cov_contribution(L, l, epsilon=0.9, theta=1.0, dt=0.5, sigma=1.0):
    if L<0 or l>L:
        return 0
    mll = np.max([L-l,l])   
    rml = np.abs(L-2*l)
    c = (sigma**2)*((theta*epsilon*dt)**L)*(1-theta*dt)**rml#*dt
    c = c * binom(mll, rml) * hyp2f1(mll+1,mll+1,rml+1,(1-theta*dt)**2)
    c = c * (1-theta*dt)**rml
    return c

def crosscov_contribution(L, l, order=1, epsilon=0.9, theta=1.0, dt=0.5, sigma=1.0):
    if order<0:
        order = -order
        l = L-l
        
    if order==0:
        c = cov_contribution(L, l, epsilon=epsilon, 
            theta=theta, dt=dt, sigma=sigma)
    elif False: #order==1:
        c = (1-theta*dt)*cov_contribution(L, l, epsilon=epsilon, 
            theta=theta, dt=dt, sigma=sigma)
        c += epsilon*theta*dt*cov_contribution(L-1, l, epsilon=epsilon, 
            theta=theta, dt=dt, sigma=sigma)
    else:
        c = (1-theta*dt#*(1+epsilon)
                )*crosscov_contribution(L, l, order=order-1, 
            epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
        c += epsilon*theta*dt*crosscov_contribution(L-1, l, order=order-1, 
            epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
    return c

def score_function(L, l, md=1, epsilon=0.9, theta=1.0, dt=0.5, sigma=1.0):
    if md==1:
        c = epsilon*theta*dt*cov_contribution(L-1, l, epsilon=epsilon, 
            theta=theta, dt=dt, sigma=sigma)
    else:
        c = crosscov_contribution(L,l, order=1, epsilon=epsilon, 
            theta=theta, dt=dt, sigma=sigma)
        for i in range(0,md+1):
            if i!=1:
                if i < 1:
                    scale = (1-theta*dt)
                else:
                    scale = (theta*epsilon*dt)**(i-1)
                c -= scale * crosscov_contribution(L,l, order=i, epsilon=epsilon, 
                    theta=theta, dt=dt, sigma=sigma)/md
    return c
```

```python
################################################################################
# get motif contributions for different parameters
dt0 = 1
edge_labels = [[0,1],[1,0],[1,1],[0,2],[2,0],[1,2],[2,1],[0,3],[3,0]]
params = [[0.9, 0.1], [0.5, 0.1], [0.9, 0.5]]
c = np.zeros((4, len(params), len(edge_labels)))
for i in range(len(params)):
    c[0,i,0] = 1-np.sign(i)*0.025
    c[1,i,:] = [ cov_contribution(el[0]+el[1],el[0], epsilon=params[i][0], 
        theta=params[i][1], dt=dt0) for el in edge_labels]
    c[2,i,:] = [ crosscov_contribution(el[0]+el[1],el[0], order=1, epsilon=params[i][0], 
        theta=params[i][1], dt=dt0) for el in edge_labels]
    c[3,i,:] = [ score_function(el[0]+el[1],el[0], md=1, epsilon=params[i][0], 
        theta=params[i][1], dt=dt0) for el in edge_labels]
    c[3,i,:] = [ score_function(el[0]+el[1],el[0], md=1, epsilon=params[i][0], 
        theta=params[i][1], dt=dt0) for el in edge_labels]
    
```

```python
def f(epsilon=0.9, theta=1.0, dt=0.5, sigma=1.0):
    #s = cov_contribution(1, 1, epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
    #s = s/cov_contribution(2, 1, epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
    s = crosscov_contribution(2, 1, order=1, epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
    s = s/cov_contribution(2, 1, epsilon=epsilon, theta=theta, dt=dt, sigma=sigma)
    
    return s
```
