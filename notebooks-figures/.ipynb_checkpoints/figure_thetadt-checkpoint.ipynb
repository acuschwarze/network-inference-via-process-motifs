```python
import sys
import matplotlib as mpl
import scipy.io as sio
from graph_models import *
from inference_methods import *

sys.path.append('../utils/')
sys.path.append('../libs/')
from utils import *

NUM_CORR = 2
NUM_NETRD = 4

mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = [r'\usepackage{amsmath}'] 

def baseline_acc(n, m):
    x = int(n ** 2 - n)
    num_t = m * (m / x)
    num_f = (x - m) * ((x - m) / x)
    return (num_t + num_f) / x

def assign_delays(adj, all_delays):
    num_edges = int(np.sum(adj))
    # Randomly assign each edge in adj a delay index 
    rand_assignments = np.random.randint(len(all_delays), size=num_edges)
    all_adj = [] 
    for k in range(len(all_delays)): # k = delay index 
        adj_k = np.copy(adj)
        assignment_ind = 0 # track which edge assignment index we are at 
        for i in range(adj_k.shape[0]):
            for j in range(adj_k.shape[1]):
                # If the original adj matrix has an edge here
                if (adj_k[i, j] == 1):
                    # And if this edge doesn't correspond with the kth delay, exclude the edge
                    if (rand_assignments[assignment_ind] != k):
                        adj_k[i, j] = 0
                    # We have encountered another edge. Increase the assignment_ind. 
                    assignment_ind += 1
        all_adj.append(adj_k)
    return all_adj 

def OU_dynamics(adj, dt, timesteps, epsilon, theta, sigma, all_delays):
    # Get number of nodes in graph 
    num_nodes = adj.shape[0]
    # Construct A
    eig_max = np.max(np.abs(np.linalg.eig(adj)[0]))
    all_adj = assign_delays(adj, all_delays)
    all_A = []
    for k in range(len(all_delays)):
        adj_k = all_adj[k]
        A_k = (epsilon / eig_max) * adj_k.T
        all_A.append(A_k)
    # Initial condition 
    x0 = np.random.normal(0, 1, size=(num_nodes))
    # Construct time series data for each node 
    X = np.empty((num_nodes, timesteps))
    X[:, 0] = x0
    for i in range(1, timesteps):
        dx = 0 
        for k, delay in enumerate(all_delays):
            if (i - delay >= 0):
                dx += all_A[k] @ X[:, i-delay]
        dx -= X[:, i-1]
        dx *= (theta * dt)
        noise = sigma * np.sqrt(dt / num_nodes) * np.random.randn(num_nodes)
        X[:, i] = X[:, i-1] + dx + noise
    return X

# def sim_no_lag(adj, epsilon, sigma, theta, dt=0.5, T=1000, x0=None, eta=0,
#     normalize_adjacency=True, scale_noise_by_time_step=True):
        
#     # factor for noise term
#     scale_noise = sigma * np.sqrt(dt/num_nodes)
#     if scale_noise_by_time_step:
#         scale_noise = scale_noise * np.exp(-theta*dt)
    
#     # Construct time series data for each node 
#     X = np.empty((num_nodes, T))
#     X[:, 0] = x0
#     for i in range(1, T):
#         dx = (theta * dt) * (A @ X[:, i-1])
#         noise = scale_noise * np.random.randn(num_nodes)
#         #noise = sigma * np.random.randn(num_nodes)
#         X[:, i] = X[:, i-1] + dx + noise
        
#     return X
```

# Accuracy across various network sizes and theta * dt values

```python
# Declare system and network parameters 
d = 0.2
epsilon = 0.9
sigma = 0.1
dt = 0.1

# Declare experiment parameters 
n_vals = [5, 10, 20, 50]
t_vals = np.linspace(0, 10, 20)
```

# Correlation Methods

```python
# Choose number of trials for correlation methods:
num_trials = 50

acc_sing = []
acc_mult = []
for i in range(NUM_CORR):
    acc_sing.append(np.empty((num_trials, len(n_vals), len(t_vals))))
    acc_mult.append(np.empty((num_trials, len(n_vals), len(t_vals))))

for k, acc in enumerate([acc_sing, acc_mult]):
    if (k == 0): # single
        print('Computing accuracy for single-delay data...')
    else: # multiple
        print('Computing accuracy for multiple-delay data...')
    
    print('Running trials...')
    for trial in range(num_trials):
        if (trial % 10 == 0):
            print('Trial ' + str(trial))
        
        for i, n in enumerate(n_vals):
            for j, theta in enumerate(t_vals):
                # Generate network 
                adj = make_adjacency(n, d)
                m = int(np.sum(adj))
                
                # Generate data 
                if (k == 0): 
                    X = sim(adj, epsilon, sigma, theta, dt=dt, max_lag=1, x0=np.zeros(n))
                    # X = OU_dynamics(adj, dt, 1000, epsilon, theta, sigma / np.sqrt(n), [1])
                else: 
                    X = sim(adj, epsilon, sigma, theta, dt=dt, max_lag=5, x0=np.zeros(n))
                    # X = OU_dynamics(adj, dt, 1000, epsilon, theta, sigma / np.sqrt(n), [1, 2, 3, 4, 5])
                
                # Cross correlation
                est1 = inf_via_crosscorrelation(X, m)
                acc[0][trial, i, j] = get_accuracy(est1, adj)
                
                # Proposed method
                if (k == 0):
                    est2 = inf_via_s(X, m, max_lag=1)
                else:
                    est2 = inf_via_s(X, m, max_lag=5)
                acc[1][trial, i, j] = get_accuracy(est2, adj) 
                
    print('Done!')
    print()
    
c_avgs_sing = []
c_avgs_mult = []
for i in range(NUM_CORR):
    c_avgs_sing.append(np.average(acc_sing[i], axis=0))
    c_avgs_mult.append(np.average(acc_mult[i], axis=0))
```

# CCM, NTE, OUI, Granger

```python
# Choose trial scaling constant (trial_const = n * num_trials)
trial_const = 250 
# Data save destination
data_dest = 'data/fig_thetadt/'

n_acc_sing = []
n_acc_mult = []

for i, n in enumerate(n_vals):
    num_trials = trial_const // n
    file_name = 'net_size_{}.mat'.format(n)
    f = sio.loadmat(data_dest + file_name)
    n_acc_sing.append([])
    n_acc_mult.append([])
    
    for k, n_acc in enumerate([n_acc_sing, n_acc_mult]):
        for j in range(NUM_NETRD):
            n_acc[i].append(np.empty((num_trials, len(t_vals))))
        
        for trial in range(num_trials):
            for j, theta in enumerate(t_vals):
                adj = f['truth'][k][trial][j]
                M1 = f['ccm'][k][trial][j]
                M2 = f['nte'][k][trial][j]
                M3 = f['oui'][k][trial][j]
                M4 = f['gc'][k][trial][j]
                m = int(np.sum(adj))
                
                # Convergent Cross Mapping
                np.fill_diagonal(M1, np.inf) 
                est1 = threshold_values(M1, m, keep_max=False)
                n_acc[i][0][trial, j] = get_accuracy(est1, adj)

                # Naive Transfer Entropy
                np.fill_diagonal(M2, -np.inf)
                est2 = threshold_values(M2, m)
                n_acc[i][1][trial, j] = get_accuracy(est2, adj)

                # OU Inference
                np.fill_diagonal(M3, -np.inf) 
                est3 = threshold_values(M3, m)
                n_acc[i][2][trial, j] = get_accuracy(est3, adj)
                
                # Granger Causality 
                np.fill_diagonal(M4, -np.inf)
                est4 = threshold_values(M4, m)
                n_acc[i][3][trial, j] = get_accuracy(est4, adj)
                
n_avgs_sing = []
n_avgs_mult = []
for j in range(NUM_NETRD):
    n_avgs_sing.append(np.empty((len(n_vals), len(t_vals))))
    n_avgs_mult.append(np.empty((len(n_vals), len(t_vals))))
for i, n in enumerate(n_vals):
    for j in range(NUM_NETRD):
        n_avgs_sing[j][i, :] = np.average(n_acc_sing[i][j], axis=0)
        n_avgs_mult[j][i, :] = np.average(n_acc_mult[i][j], axis=0)
```

# Baseline

```python
rand_acc = np.empty((len(n_vals)))
for i, n in enumerate(n_vals):
    m = int(d * n * (n - 1))
    rand_acc[i] = baseline_acc(n, m)
```

```python
all_avgs_sing = c_avgs_sing[:] + n_avgs_sing[:]
all_avgs_mult = c_avgs_mult[:] + n_avgs_mult[:]

labels = []
labels.append(r'$\Sigma$')
labels.append(r'$\mathrm{S}$')
labels.append(r'$\mathrm{CCM}$')
labels.append(r'$\mathrm{NTE}$')
labels.append(r'$\mathrm{OUI}$')
labels.append(r'$\mathrm{G}$')

# Make plot
plt.figure(figsize=(15, 8), dpi=200)

for k, all_avgs in enumerate([all_avgs_sing, all_avgs_mult]):
    for i, n in enumerate(n_vals):
        ax = plt.subplot(2, len(n_vals), i + (k * len(n_vals)) + 1)
        
        for j, accuracy in enumerate(all_avgs):
            if (j == 1):
                plt.plot(t_vals*dt, accuracy[i, :], label=labels[j], linewidth=2, marker='o')
            else:
                plt.plot(t_vals*dt, accuracy[i, :], label=labels[j], linewidth=2, marker='o', linestyle='dashed')
                
        plt.plot(t_vals*dt, rand_acc[i]*np.ones((len(t_vals))), '--', label=r'$\mathrm{Baseline}$', linewidth=2, c='k')

        if (k == 0): # single
            plt.title(r'$n = ' + str(n) + '$', fontsize=20)
            ax.set_xticklabels([])
        else: # multiple
            plt.xlabel(r'$\theta \Delta t$', fontsize=20)
            
        plt.xticks(size=15)
        plt.yticks(size=15)
        if (i == 0):
            plt.ylabel(r'$\mathrm{Accuracy}$', fontsize=20)
        else:
            ax.set_yticklabels([])
        plt.ylim([0.45, 1.05])
    
plt.legend(fontsize=12, ncol=2)
plt.tight_layout()
plt.show()
```
