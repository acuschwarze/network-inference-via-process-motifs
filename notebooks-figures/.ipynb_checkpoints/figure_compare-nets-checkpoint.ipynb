# Transitivity

```python
x = Variable('n', np.arange(30, 80, 10))
y = Variable('density', [0.1, 0.2])
z = Variable('network_type', ['ER','BA','RL','RR','SW'])
num_trials = 10

transitivity = np.empty((len(x.values), len(y.values), len(z.values), num_trials))
accuracy = np.empty((len(x.values), len(y.values), len(z.values), num_trials))

for trial in range(num_trials):
    for i, _n in enumerate(x.values):
        for j, _density in enumerate(y.values):
            for k, _network_type in enumerate(z.values):
                res = siminf(network_type=_network_type, n=_n, density=_density, inference_method='SRC')
                G = nx.from_numpy_matrix(res['adjacency'], create_using=nx.DiGraph)
                # transitivity[i, j, k, trial] = nx.transitivity(G)
                transitivity[i, j, k, trial] = get_spectral_radius(res['adjacency'])
                accuracy[i, j, k, trial] = res['accuracy']

# Plot
y_markers = ['o', 's']
z_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
legend_elements = []
for j, density in enumerate(y.values):
    legend_elements.append(Line2D([0], [0], marker=y_markers[j], label='density={}'.format(density), 
                                  c='w', mec='k', markerfacecolor='w', markersize=10))
for k, network_type in enumerate(z.values):
    legend_elements.append(Line2D([0], [0], marker='s', label=network_type, c='w',
                                  markerfacecolor=z_colors[k], markersize=10))
plt.figure(dpi=100)
for trial in range(num_trials):
    for i, n in enumerate(x.values):
        for j, density in enumerate(y.values):
            for k, network_type in enumerate(z.values):
                plt.scatter(transitivity[i, j, k, trial], accuracy[i, j, k, trial], 
                            s=50, marker=y_markers[j], c=z_colors[k], edgecolor='k')
# plt.xlabel('Transitivity')
plt.xlabel('Max Eig')
plt.ylabel('Accuracy')
plt.legend(handles=legend_elements, loc=(1.05, 0))
plt.show()
```

```python
t = time.time()
x = Variable('n', [50]*10)
y = Variable('density', [0.1, 0.2])
z = Variable('network_type', ['ER','BA','RL','RR','SW'])
num_trials = 100

shape = (len(x.values), len(y.values), len(z.values), num_trials)
transitivity = np.empty(shape)
accuracy = np.empty(shape)
kmaxin = np.empty(shape)
kmaxout = np.empty(shape)
kstdin = np.empty(shape)
kstdout = np.empty(shape)

for trial in range(num_trials):
    for i, _n in enumerate(x.values):
        for j, _density in enumerate(y.values):
            for k, _network_type in enumerate(z.values):
                res = siminf(network_type=_network_type, n=_n, density=_density, inference_method='SRC')
                G = nx.from_numpy_matrix(res['adjacency'], create_using=nx.DiGraph)
                #print(list(G.in_degree()))
                kmaxin[i,j,k,trial] = max(dict(G.in_degree()).values())
                kmaxout[i,j,k,trial] = max(dict(G.out_degree()).values())
                kstdin[i,j,k,trial] = np.std(list(dict(G.in_degree()).values()))
                kstdout[i,j,k,trial] = np.std(list(dict(G.out_degree()).values()))                
                transitivity[i, j, k, trial] = nx.transitivity(G)
                get(res['adjacency'])
                accuracy[i, j, k, trial] = res['accuracy']

# Plot
for ix, xvar in enumerate([transitivity, kmaxin, kmaxout, kstdin, kstdout]):
    y_markers = ['o', 's']
    z_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
    legend_elements = []
    for j, density in enumerate(y.values):
        legend_elements.append(Line2D([0], [0], marker=y_markers[j], label='density={}'.format(density), 
                                      c='w', mec='k', markerfacecolor='w', markersize=10))
    for k, network_type in enumerate(z.values[:1]):
        legend_elements.append(Line2D([0], [0], marker='s', label=network_type, c='w',
                                      markerfacecolor=z_colors[k], markersize=10))
        
    for j, density in enumerate(y.values):
        plt.figure(dpi=100)
        for trial in range(num_trials):
            for i, n in enumerate(x.values):    
                for k, network_type in enumerate(z.values[:1]):
                    plt.scatter(xvar[i, j, k, trial], accuracy[i, j, k, trial], 
                                s=50, marker=y_markers[j], c=z_colors[k], edgecolor='k')
        plt.xlabel(['Transitivity', 'max in-degree', 'max out-degree', 'std in-degree', 'std out-degree'][ix])
        plt.ylabel('Accuracy')
        plt.legend(handles=legend_elements, loc=(1.05, 0))
        plt.show()
        
for ix, xvar in enumerate([transitivity, kmaxin, kmaxout, kstdin, kstdout]):
    y_markers = ['o', 's']
    z_colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red', 'tab:purple']
    legend_elements = []
    for j, density in enumerate(y.values):
        legend_elements.append(Line2D([0], [0], marker=y_markers[j], label='density={}'.format(density), 
                                      c='w', mec='k', markerfacecolor='w', markersize=10))
    for k, network_type in enumerate(z.values):
        legend_elements.append(Line2D([0], [0], marker='s', label=network_type, c='w',
                                      markerfacecolor=z_colors[k], markersize=10))
        
    for j, density in enumerate(y.values):
        plt.figure(dpi=100)
        for trial in range(num_trials):
            for i, n in enumerate(x.values):    
                for k, network_type in enumerate(z.values):
                    plt.scatter(xvar[i, j, k, trial], accuracy[i, j, k, trial], 
                                s=50, marker=y_markers[j], c=z_colors[k], edgecolor='k')
        plt.xlabel(['Transitivity', 'max in-degree', 'max out-degree', 'std in-degree', 'std out-degree'][ix])
        plt.ylabel('Accuracy')
        plt.legend(handles=legend_elements, loc=(1.05, 0))
        plt.show()
print(time.time()-t)
```
