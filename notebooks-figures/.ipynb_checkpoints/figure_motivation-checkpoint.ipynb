```python
# set parameters
default_pars = lambda : {'net_size': 10, 'density': 0.5, 'directedness': 0.66, 
                         'sigma': 0.2, 'epsilon': 0.9, 'theta': 0.5, 'dt': 1}
num_trials = 100

# set parameter ranges
dx = 0.1
dn = 1
n0 = 5
methods = ['base', 'GC', 'S', 'CCO', 'CCM', 'NTE', 'OUI']
xvals = [np.linspace(0.1,0.9,11), # density
         np.linspace(0,1,11), # directedness
         1/np.linspace(1,100,11), # autocorrelation
         np.linspace(1,12,11, dtype=int), # lag
         np.linspace(5,15,num=11, dtype=int) # computation time
         ]

print([len(x) for x in xvals])

# create or load data
load = True
save = True
data = np.zeros((5, len(methods), len(xvals[0]), num_trials))

t0 = time.time()
for im, m in enumerate(methods):
    file = '../data/figure_0-'+m+'.data'
    load_m = load
    if load_m==True:
        if os.path.exists(file): #try:
            data[:,im, :, :] = pickle.load(open(file, 'rb'))
            print('Load data for', m)
        else:
            load_m = False
            print('No files found.')
            
    if load_m==False:
        print('Run simulations. This can take some time.')
        for i, kws in enumerate([['density'], ['directedness'], ['theta'], ['lag', 'max_lag'], ['n']]):
            print(i)
            if 'n' in kws:
                prop = 'time'
            else:
                prop = 'accuracy'
            pars = default_pars()  
            pars['inference_method'] = m
            if i==0:
                pars['directedness'] = 0
            for ix, x in enumerate(xvals[i]):
                for kw in kws:
                    pars[kw] = x
                for t in range(num_trials):
                    data[i,im,ix,t] = siminf(**pars)[0][prop]

        if save:
            pickle.dump(data[:,im,:,:], open(file, 'wb'))
            
print(time.time()-t0)
```

```python
xlabels = ['Density','Directedness','Autocorrelation time scale','Network size']
# ['density', 'directedness', 'theta', 'n']
ylabels = ['Accuracy', 'Accuracy', 'Accuracy', 'Accuracy','Computation time (ms)']
linelabels = ['baseline', 'Granger', 'C4',  'CCO', 'CCM', 'NTE', 'OUI']
linelabels = methods

# rearrange data and labels for subplots
if True:
    f = lambda arr: [arr[0], arr[1], arr[3], arr[2]]
    xvals = f(xvals)
    xlabels = f(xlabels)
    ylabels = f(ylabels)
    data = np.array(f(data))
```

```python
color3 = ['#022957','#ff8000','#6e6e6e']
markers = []

plt.figure(figsize=(4,4)) 
for i in range(4):
    ax = plt.subplot(2,2,1+i)
    plt.xlabel(xlabels[i])
    plt.ylabel(ylabels[i])
    
    if xlabels[i]=='Autocorrelation time scale':
        for im, m in enumerate(methods):
            plt.plot(1/xvals[i], np.mean(data[i,im,:,:], axis=-1), 
                marker='.', #color=color3[im], 
                label=linelabels[im])
            
    elif ylabels[i]=='Computation time (ms)':
        for im, m in enumerate(methods):
            plt.plot(xvals[i], 1000*np.mean(data[i,im,:,:], axis=-1), 
                marker='.', #color=color3[im], 
                label=linelabels[im])             
    else:
        for im, m in enumerate(methods):
            plt.plot(xvals[i], np.mean(data[i,im,:,:], axis=-1), 
                marker='.', #color=color3[im], 
                label=linelabels[im])
    if i==0:
        plt.legend(bbox_to_anchor=(1.26, 1.3), ncol=3, loc='upper center')
        
    #if i % 2:
    #    ax.yaxis.tick_right()
    #    ax.yaxis.set_label_position('right')

plt.subplots_adjust(hspace=0.4,wspace=0.5)
plt.savefig('../figures/figure_motivation.pdf')
```

```python
def grey2hex(x):
    x = int(x)
    x = min([x,255])
    x = max([0,x])
    c = '#%02x%02x%02x' % (x,x,x)
    return c
```
