In the figure below, we show how accuracy changes when increasing number of samples. In one subplot, we keep overall time fixed. In the other subplots, we keep the sampling rate fixed. This figure appears in Section 6D of our paper.

```python
################################################################################
# In this figure, we show how accuracy changes when increasing number of 
# samples. In one subplot, we keep overall time fixed. In the other subplot, we
# keep the sampling rate fixed. This figure appears in Section 6D of our paper.

import warnings
warnings.filterwarnings('ignore')
num_trials=100
values = np.arange(45,205,5) 
values = np.arange(5,205,5)
values = np.arange(100,1050,50)

methods = ['base','OUI','GC','TE','CM','LC','LCRC','LCCF']
linecolors=['k','#c0c0c0','#a0a0a0','#808080','#606060','#404040', 'salmon', 
            'tab:red']
#linecolors=['red','blue','purple','green', 'yellow', 'orange', 'black','gray']
linestyles=['--', ':', ':', ':', ':', ':', '-', '-']
linewidths=[1, 1, 1, 1, 1, 1, 1, 1]
markers=['.', '^', 'd', 'p', '*', 'x','D', 's']
markersizes=[5, 7, 7, 5, 5, 5, 3, 3]

if True:
    # We cannot generate data for fixed observation times case using the plot1d
    # function because there is no implementation for fixed observation times
    # in that function. Instead, we generate the data for the fixed-
    # observation-times case manually. (We later add it to the figure.)
    datalist = [[],[]]
    for ir, rec in enumerate([1.0, None]):
        for me in methods:
            filename = ('../data/figure_D1_if-'+me+str(num_trials)
                +'_rec'+str(rec)+'_T'+array2string(values, abbrv=True)+'.npy')
            load_failed = True
            if load:
                if os.path.exists(filename):
                    print('load data from file for '+me+'...', end='')
                    a = np.load(filename)
                    if a.shape[1] >= num_trials:
                        a = a[:,:num_trials]
                        load_failed = False
            if load_failed: # or me=='LCCF':
                print('compute data for '+me+'...', end='')
                # check if this line is correct
                pars = [{#'epsilon':0.4, 
                         'reciprocity':rec,
                         'inference_method':me, 'T':N, 'dt':100/N} for N in values]
                a = np.zeros((len(values), num_trials))
                for i, p in enumerate(pars):
                    a[i] = [siminf(**p)['accuracy'] for _ in range(num_trials)]
                np.save(filename, a)
            datalist[ir] += [a]
        print('')

#num_trials=100
fig, axes, legend = plot1d(nrows=1, ncols=3, y='accuracy', 
    num_trials=num_trials, x=Variable('T',values), fheight=4,
    #linevars=Variable('inference_method', methods),
    linevars=Variable('inference_method', methods),
    linevar_labels=methods, default_parameters = {'reciprocity':None},
    add_subplotlabels=False, subplotlabelx=0.06, subplotlabely=0.9,                       
    linecolors=linecolors, linestyles=linestyles, linewidths=linewidths, 
    load=load, save=True, draw=draw, add_legend=False, 
    markers=markers, markersizes=markersizes, sharey=False, alpha=0.5)

if draw:
    # Currently both subplots show the same data. Now we replace the data in 
    # subplot 2 with the data for fixed observation times, which we generated
    # manually above.
    num_lines = len(axes[1].lines)
    for i, line in enumerate(reversed(axes[1].lines)):
        axes[1].lines.pop(num_lines-i-1)
        del line

    for i in range(len(methods)):
        data = datalist[1][i]
        axes[1].plot(values, np.mean(data, axis=-1), c=linecolors[i], 
                     ls=linestyles[i], lw=linewidths[i],
                     marker=markers[i], ms=markersizes[i], label=methods[i])
        
    num_lines = len(axes[2].lines)
    for i, line in enumerate(reversed(axes[2].lines)):
        axes[2].lines.pop(num_lines-i-1)
        del line

    for i in range(len(methods)):
        data = datalist[0][i]
        axes[2].plot(values, np.mean(data, axis=-1), c=linecolors[i], 
                     ls=linestyles[i], lw=linewidths[i],
                     marker=markers[i], ms=markersizes[i], label=methods[i])

    # make figure pretty
    fig.set_size_inches([8,3])
    for i in range(3):
        axes[i].set_xlabel('observations $N$')
        axes[i].set_ylim([0.485, 0.905])
        axes[i].set_yticks(np.arange(0.5, 0.95, 0.05))
        
    for i in range(1,3):
        axes[i].set_ylabel('')
        axes[i].set_yticklabels([])
        ypos = [0, 0.7, 0.35][i]
        axes[i].text(0.65-0.025, ypos, r'$\Delta t_\tau\approx0.15$', 
            color='blue', zorder=-6, fontsize=9, backgroundcolor='white', 
            transform=axes[i].transAxes)
        axes[i].axvline(x=650, ls=":", zorder=-5, color='blue')
        axes[i].text(0.25-0.025, 0.8, r'$\Delta t_\tau=0.4$', 
            color='magenta', zorder=-4, fontsize=9, backgroundcolor='white', 
            transform=axes[i].transAxes)
        axes[i].axvline(x=250, ls=":", zorder=-3, color='magenta')
        axes[i].text(0.10-0.025, 0.905, r'$\Delta t_\tau=1$', 
            color='purple', zorder=-2, fontsize=9, backgroundcolor='white', 
            transform=axes[i].transAxes)
        axes[i].axvline(x=100, ls=":", zorder=-1, color='purple')

        axes[i].text(0.875, 0.88, '('+['a','b','c'][i]+')', 
            transform=axes[i].transAxes)
        
    axes[0].text(0.05, 0.88, '(a)', transform=axes[0].transAxes)
    
    axes[0].legend(ncol=2, fontsize=9,
        bbox_to_anchor=(1.0, 0.46), 
        loc='upper right', columnspacing=1)
    
    axes[0].set_title(r'Fixed sampling rate, $r_e=0.5$',fontsize=11)
    axes[1].set_title(r'Fixed observation time, $r_e=0.5$',fontsize=11)
    axes[2].set_title(r'Fixed observation time, $r_e=1$', fontsize=11)
    
    plt.subplots_adjust(left=0.07, right=0.99, top=0.91, bottom=0.15, 
                        hspace=0.225, wspace=0.1)    

    plt.savefig('../figures/resultsD1.pdf')
```
