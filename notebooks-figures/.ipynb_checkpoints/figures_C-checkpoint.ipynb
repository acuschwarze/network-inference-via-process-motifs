# Figures for Network Inference from Noisy Data

Start by importing some packages and defining some functions.

```python
import sys, time
sys.path.append('../utils/')
from utils import *
import matplotlib.pyplot as plt  

################################################################################
# This block checks whether code is run in a jupyter notebook or as script. We
# use the script to generate data (`load=False`) and the notebooks to draw
# figures (`draw=True`).

if is_interactive():
    print('is_interactive', is_interactive())
    draw = True
    load = True
    num_trials = 100
    print('num_trials', num_trials)
    rcPhysRev(fontsize=10)
else:
    print('is_interactive', is_interactive())
    draw = False
    load = False
    num_trials = 100
    print('num_trials', num_trials)
```

The figure below shows accuracy as a function of noise parameters of the stochastic difference model. You can find it in Section 6C of our paper.

```python
################################################################################
# This cell makes a figure to show accuracy as a function of noise parameters of 
# the stochastic difference model. You can find it in Section 6C of our paper.

nolag = 1
alag = 5+1

t=time.time()
fig, axes, legend = plot1d(ncols=2, nrows=2, num_trials=num_trials, 
    xvars=[Variable('sigma', [10**(i) for i in range(-5,3)]),
        Variable('sigma', [10**(i) for i in range(-5,3)]),
        Variable('eta', [10**(i) for i in range(-5,3)]),
        Variable('eta', [10**(i) for i in range(-5,3)])],
    plotvars=Variable('max_lag',[nolag, alag, nolag, alag]),
    default_parameters={}, load=draw, save=True, verbose_sim=False,
    linevars=Variable('inference_method',
                      ['base','OUI','GC','TE','CM','LC','LCRC','LCCF']),
    linevar_labels=['base','OUI','GC','TE','CM','LC','LCRC','LCCF'],
    add_subplotlabels=True, subplotlabelx=0.86, subplotlabely=0.875,                       
    #linevars=Variable('inference_method',['LIC']),
    linecolors=['k','#c0c0c0','#a0a0a0','#808080','#606060','#404040' ,
                'salmon', 'tab:red'],
    sharey=False, add_legend=False, fheight=4.9,
    linestyles=['--', ':', ':', ':', ':', ':', '-', '-'], 
    linewidths=[1, 1, 1, 1, 1, 1, 1, 1], 
    markers=['.', '^', 'd', 'p', '*', 'x','D', 's'], markersizes=[5]*8)

for i in range(2):
    for j in range(2):
        axes[i,j].set_xscale('log')
        axes[i,j].set_xticks([0.0001, 0.01, 1, 100])
    axes[0,i].set_ylim([0.48,1.02])
    axes[0,i].set_yticks([0.5, 0.6, 0.7, 0.8, 0.9, 1.0])
    axes[1,i].set_ylim([0.48,0.92])
    axes[1,i].set_yticks([0.5, 0.6, 0.7, 0.8, 0.9])
    axes[1,i].axvline(x=0.2, ls=":", zorder=-1, color='magenta')
    axes[1,i].text(0.36, 0.15, r'$\eta=\sigma$', color='magenta', 
                   fontsize=10, transform=axes[1,i].transAxes)
    axes[i,1].set_ylabel('')
    axes[i,1].set_yticklabels([])
axes[0,0].set_title(r'$\delta=0$')
axes[0,1].set_title(r'$\delta=5$')

if True:
    plt.sca(axes[1,0])
    le = plt.legend(#list(axes[2,0].lines)[::-1], 
        #[ll.get_label() for ll in list(axes[2,0].lines)[::-1]],
        ncol=4, fontsize=9,
        bbox_to_anchor=(1.03, -0.28), 
        loc='upper center')#  loc=(1.1, 0))
    plt.subplots_adjust(left=0.16, right=0.99, top=0.95,
                        bottom=0.2, hspace=0.35, wspace=0.15)

plt.savefig('../figures/resultsC1.pdf')
#print(time.time()-t)
```

Below, we show the same figure as the one above, but *without* a log scale for the x axis. We do not use this figure in the paper.

```python
################################################################################
# This cell makes a similar figure to the one above, but *without* a log scale
# for the x axis. We do not use this figure in the paper.

nolag = 1
alag = 5+1
t=time.time()
fig, axes, legend = plot1d(ncols=2, nrows=2, num_trials=num_trials, 
    xvars=repeatList(['sigma','eta'],2),
    plotvars=Variable('max_lag', tileList([nolag, alag], 2)),
    default_parameters={}, load=draw, save=True, verbose_sim=False,
    linevars=Variable('inference_method',
                      ['base','OUI','GC','TE','CM','LC','LCRC','LCCF']),
    linecolors=['k','#c0c0c0','#a0a0a0','#808080','#606060','#404040' ,
                'salmon', 'tab:red'],
    sharey=False, add_legend=False, fheight=4.9,
    linestyles=['--', ':', ':', ':', ':', ':', '-', '-'], 
    linewidths=[1, 1, 1, 1, 1, 1, 1, 1], 
    markers=['.', '^', 'd', 'p', '*', 'x','D', 's'], markersizes=[5]*8)

for i in range(2):
    for j in range(2):
        axes[i,j].set_ylim([0.48,0.92])
        axes[i,j].set_yticks([0.5, 0.6, 0.7, 0.8, 0.9])
        axes[i,j].set_ylim([0.48,0.92])
    axes[i,1].set_ylabel('')
    axes[i,1].set_yticklabels([])
axes[0,0].set_title(r'$\delta=0$')
axes[0,1].set_title(r'$\delta=5$')

if True:
    plt.sca(axes[1,0])
    le = plt.legend(#list(axes[2,0].lines)[::-1], 
        #[ll.get_label() for ll in list(axes[2,0].lines)[::-1]],
        ncol=4, fontsize=9, bbox_to_anchor=(1.03, -0.28), 
        loc='upper center')#  loc=(1.1, 0))
    plt.subplots_adjust(left=0.16, right=0.99, top=0.95,
                        bottom=0.2, hspace=0.35, wspace=0.15)

plt.savefig('../figures/resultsC2.pdf')
#print(time.time()-t)
```
