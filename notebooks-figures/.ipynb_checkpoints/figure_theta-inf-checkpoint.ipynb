```python
import sys
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
sys.path.append('../utils/')
from utils import *
import time

rcPhysRev(fontsize=10)
```

```python
xvars = [Variable('n', np.arange(5, 55, 5)), 
         Variable('density', np.arange(0.1, 1, 0.1)),
         Variable('directedness', np.arange(0, 1.1, 0.1)), 
         Variable('epsilon', np.arange(0, 1, 0.1)), 
         Variable('sigma', np.arange(0.5, 5.5, 0.5))]
num_trials = 100

plot_data = []
for x in xvars:
    print('Computing for ' + x.name + '...')
    theta_acc = np.zeros((len(x.values), num_trials))
    for it in range(num_trials):
        for ix, xval in enumerate(x.values):
            pars = default_siminf_pars()
            pars[x.name] = xval
            theta_inf,_ = infer_parameters(siminf(**pars)['sim'], pars['dt'])
            theta_acc[ix,it] = np.linalg.norm(theta_inf-pars['theta'])/np.linalg.norm(pars['theta'])
    plot_data.append(np.average(theta_acc, axis=-1))
```

```python
plt.figure(dpi=100)
for i, x in enumerate(xvars):
    plt.subplot(2, 3, i + 1)
    plt.plot(x.values, plot_data[i], c=c8(i), linewidth=2, marker='o', markersize=4)
    plt.xlabel(get_labelstring(x.name))
    if (i % 3 == 0):
        plt.ylabel('Theta Relative Error')
    # plt.ylim([0.01, 0.06])
plt.tight_layout()
plt.show()
```
