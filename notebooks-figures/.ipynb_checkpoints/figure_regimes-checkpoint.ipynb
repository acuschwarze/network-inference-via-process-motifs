#### Define some functions for drawing figure elements ...

```python
def poly_at_angle(angle):
    if angle >=45:
        rad = (angle)*np.pi/180.0
        xvals = np.array([0.0, np.sin(np.pi/2-rad), 1.0])
        yvals1 = np.array([0.0, 1.0, 1.0])
        yvals2 = np.array([0.0, 0.0, 0.0])
    else:
        rad = angle*np.pi/180.0
        xvals = np.array([0.0, 1.0])
        yvals1 = np.array([0.0, np.sin(rad)])
        yvals2 = np.array([0.0, 0.0])
        
    return (xvals, yvals1, yvals2)
```

#### Draw Figures ...
(Version 2 of the figures below is Figure 3 in our paper.)

Draw version 1 ...

```python
#patterns = ('-', '+', 'x', '\\', '*', 'o', 'O', '.')
colors = ['#888888', '#0331a1', '#007fbd', '#5bc7e1', '#c901a1', '#710b79', '#f88810']
blues5 = ['#03045e', '#0077b6', '#00b4d8', '#90e0ef', '#caf0f8']
blues_interpolate12 = ['#888888', '#899293', '#8A9C9F', '#8BA5AA', '#8CAFB6', '#8CB9C1', '#8DC3CD', '#8ECCD8','#8FD6E4','#90E0EF']
blues_interpolate34 = ['#00B4D8', '#00A0CA', '#018DBD', '#0179AF', '#0166A2', '#025294', '#023F87', '#022B79', '#03186C', '#03045E']
dx = 0.5
dx2=0.055

plt.figure(figsize=(4,4))
ax = plt.subplot(111)
ax.set_xlim([0,1])
ax.set_ylim([0,1])
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel(r'Characteristic timescale $\tau$')
ax.set_ylabel(r'Inverse sampling rate $\Delta t$')

# first area
ax.fill_between([0.0, 1.0], [1.0, 1.0], [0.0, 0.0], 
                facecolor=colors[0], label='n/a')

# transition and second area
num_blues = len(blues_interpolate12)
offset = int((num_blues-1)/2)/2
for i in range(num_blues-1):
    ax.fill_between(*poly_at_angle(66+offset-i/2), 
                    facecolor=blues_interpolate12[i+1],
                    label=('VAR' if i==num_blues-2 else None))

# third area
ax.fill_between([0.0, 1.0], [0.0, 1.0], [0.0, 0.0], 
                facecolor=blues5[2], label='SDD')

# transition and fourth area
num_blues = len(blues_interpolate34)
offset = int((num_blues-1)/2)/2
for i in range(num_blues-1):
    ax.fill_between(*poly_at_angle(45-21+offset-i/2), 
                    facecolor=blues_interpolate34[i+1],
                    label=('OUP' if i==num_blues-2 else None))

# legend
plt.legend(loc=1)

# middle line
ax.plot([0,0.5-dx2], [0,0.5-dx2], 'k--')
ax.plot([0.5+dx2,1], [0.5+dx2,1], 'k--')
ax.text(0.5, 0.5, 
        r'$\Delta t=\tau$', 
        rotation=45, fontsize=9, 
        horizontalalignment='center', verticalalignment='center')

# annotations
ax.text(0.0225, 0.85, 
        'very fast reverting\nprocess (and/or\ngreatly under-\nsampled\nprocess)', 
        rotation=0, fontsize=9, 
        horizontalalignment='left', verticalalignment='center')
ax.text(0.45, 0.625, 
        'fast reverting process\n(and/or\, slightly undersampled process)', 
        rotation=45, fontsize=9, 
        horizontalalignment='center', verticalalignment='center')
ax.text(0.625, 0.45, 
        'slowly reverting process\n(and/or\, slightly oversampled process)', 
        rotation=45, fontsize=9, 
        horizontalalignment='center', verticalalignment='center')
ax.text(0.625, 0.075, #0.8, 0.2, 
        'very slowly reverting process\n(and/or\, greatly oversampled process)', 
        rotation=0, fontsize=9, color='white',
        horizontalalignment='center', verticalalignment='center')
ax.text(-0.02,-0.02,'0', fontsize=9, 
        horizontalalignment='right', verticalalignment='top')

# savefig
plt.savefig('../figures/figure_regimes_v2.pdf')
```
