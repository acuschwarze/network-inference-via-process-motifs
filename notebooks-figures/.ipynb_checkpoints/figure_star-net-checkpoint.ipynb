```python
import sys
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
import seaborn as sns
sys.path.append('../utils/')
from utils import *
import time
rcPhysRev(fontsize=10)

def generate_double_star_graph(n, v1_in_deg, v2_in_deg, return_graph_params=False, 
                               center_dist=1, r=0.5, pallet=None):
    if (n < 2):
        raise ValueError('n is too small.')
    
    # Create directed graph with n nodes
    G = nx.DiGraph()
    G.add_nodes_from(['{}'.format(i + 1) for i in range(n)])

    # Add 2-way connection between center nodes v1 and v2
    G.add_edge('1', '2')
    G.add_edge('2', '1')
    v1_nodes = list(G.nodes())[2:(n//2)+1] # Nodes surrounding v1
    v2_nodes = list(G.nodes())[(n//2)+1:]  # Nodes surrounding v2

    # Nodes surrounding v1 either have a connection to v1 OR from v1
    v1_in = v1_nodes[:v1_in_deg-1]
    v1_out = v1_nodes[v1_in_deg-1:]
    G.add_edges_from([(v1_in_node, '1') for v1_in_node in v1_in])
    G.add_edges_from([('1', v1_out_node) for v1_out_node in v1_out])

    # Nodes surrounding v2 either have a connection to v2 OR from v2
    v2_in = v2_nodes[:v2_in_deg-1]
    v2_out = v2_nodes[v2_in_deg-1:]
    G.add_edges_from([(v2_in_node, '2') for v2_in_node in v2_in])
    G.add_edges_from([('2', v2_out_node) for v2_out_node in v2_out])
    
    if not return_graph_params:
        return G
    
    # Get node positions
    v1_coord = (-center_dist / 2, 0)
    v2_coord = (center_dist / 2, 0)
    angle = (2 * np.pi) / (n / 2)
    pos = {'1' : v1_coord, '2' : v2_coord}
    for k, v in enumerate(v1_nodes):
        x = (r * np.cos((k + 1) * angle)) + v1_coord[0]
        y = (r * np.sin((k + 1) * angle)) + v1_coord[1]
        pos[v] = (x, y)
    for k, v in enumerate(v2_nodes):
        x = (r * np.cos(-(((k + 1) * angle) + np.pi))) + v2_coord[0]
        y = (r * np.sin((k + 1) * angle)) + v2_coord[1]
        pos[v] = (x, y)
        
    # Get node colors
    if pallet is None:
        pallet = ['tab:blue','tab:orange','tab:green','tab:red','tab:purple','tab:pink']
    color_map = {'1' : pallet[0], '2' : pallet[3]}
    for v in v1_in:
        color_map[v] = pallet[1]
    for v in v1_out:
        color_map[v] = pallet[2]
    for v in v2_in:
        color_map[v] = pallet[4]
    for v in v2_out:
        color_map[v] = pallet[5]
    colors = [color_map[v] for v in G.nodes()]
    
    return G, pos, colors
```

```python
n = 22
v1_in_degs = np.arange(1, (n // 2) + 1)
v2_in_degs = np.arange(1, (n // 2) + 1)
v1_in_plot = np.append(v1_in_degs, v1_in_degs[-1] + (v1_in_degs[1] - v1_in_degs[0]))
v2_in_plot = np.append(v2_in_degs, v2_in_degs[-1] + (v2_in_degs[1] - v2_in_degs[0]))

acc_src = np.empty((len(v1_in_degs), len(v2_in_degs)))
for v1_ind, v1_in_deg in enumerate(v1_in_degs):
    for v2_ind, v2_in_deg in enumerate(v2_in_degs):
        G = generate_double_star_graph(n, v1_in_deg, v2_in_deg)
        acc_src[v1_ind, v2_ind] = siminf(A=get_adjacency(G), inference_method='SRC')['accuracy']
```

# Another star graph

```python
import itertools

def generate_single_star_graph(n, pd, create_all=True):
    
    num_directed = int(pd*n)
    Gs = []
    for c in itertools.combinations(range(1,int(n)+1), num_directed):
        #print(n)
        G = nx.star_graph(int(n))
        G = nx.DiGraph(G)
        #print(G.edges())
        #print(c)
        for neighbor in c:
            if np.random.uniform()<0.5:
                G.remove_edge(0, neighbor)
            else:
                G.remove_edge(neighbor, 0)
        Gs += [nx.DiGraph(G)]
        if not create_all:
            return G
        
    return Gs
    
```

```python
num_neighbors = np.arange(5,25,5)
acc = np.zeros((len(num_neighbors), max(num_neighbors)))
for i, n in enumerate(num_neighbors):
    print(i,n)
    for j, pd in enumerate((np.arange(n)+1)/n):
        Gs = generate_single_star_graph(n, pd, create_all=True)
        acc[i,j] = np.nanmean([siminf(A=get_adjacency(G), inference_method='SRC')['accuracy'] for G in Gs])   
```

```python
for i in range(len(num_neighbors)):
    plt.plot(acc[i,:num_neighbors[i]], label=num_neighbors[i])
```
