```python
import sys, pickle, time, os
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import networkx as nx

sys.path.append('../utils/')
sys.path.append('../libs/')
from utils import rcPhysRev, c8, siminf, get_adjacency, get_spectral_radius
```

```python
# settings for pretty plots
rcPhysRev()
```

```python
# set parameters
default_pars = lambda : {'net_size': 10, 'density': 0.5, 'directedness': 0.5, 
                         'sigma': 0.2, 'epsilon': 0.9, 'theta': 0.5, 'dt': 1}
num_trials = 100

# set parameter ranges
dx = 0.1
dn = 1
n0 = 5
xnum = 11
methods = ['S', 'GC', 'CCM']#['base', 'GC', 'S', 'CCO', 'CCM', 'NTE', 'OUI']
xvar_sets = [['density'], ['directedness'], ['theta'], ['lag', 'max_lag'], ['n']]
xvals = [np.linspace(0.1,0.9,num=xnum), # density
         np.linspace(0,1,num=xnum), # directedness
         1/np.linspace(1,101,num=xnum), # autocorrelation
         np.linspace(1,11,num=xnum, dtype=int), # lag
         np.linspace(n0*1,n0*11,num=xnum, dtype=int) # computation time
         ]

# create or load data
load = True
save = True
ACCU = {} #np.zeros((5, len(methods), len(xvals[0]), num_trials))
TIME = {} #np.zeros((5, len(methods), len(xvals[0]), num_trials))

t0 = time.time()
for im, m in enumerate(methods):
    ACCU[m] = {}
    TIME[m] = {}
    for ix, xvars in enumerate(xvar_sets):
        file = '../data/figure_issues-'+m+'_'+xvars[0]+'.data'
        load_m = load
        if load_m==True:
            if os.path.exists(file): 
                ACCU[m][xvars[0]], TIME[m][xvars[0]] = pickle.load(open(file, 'rb'))
                print('Load data for', m, xvars)
            else:
                load_m = False
                print('No files found.')
            
        if load_m==False:
            print('Run simulations for',m, xvars, '. This can take some time.')
            t1 = time.time()
            pars = default_pars()  
            pars['inference_method'] = m
            if 'density' in xvars:
                pars['directedness'] = 0
                
            ACCU[m][xvars[0]] = np.zeros((len(xvals[ix]),num_trials))
            TIME[m][xvars[0]] = np.zeros((len(xvals[ix]),num_trials))
            
            for iv, x in enumerate(xvals[ix]):
                for xvar in xvars:
                    pars[xvar] = x
                    for t in range(num_trials):
                        res = siminf(**pars)[0]
                        ACCU[m][xvars[0]][iv,t] = res['accuracy']
                        TIME[m][xvars[0]][iv,t] = res['time']

            if save:
                pickle.dump([ACCU[m][xvars[0]], TIME[m][xvars[0]]], open(file, 'wb'))
            print('Completed in', time.time()-t1)
            
print(time.time()-t0)
```

```python
plt.figure(figsize=(4,8))
methods = ['S', 'GC', 'CCM']
num_plots = len(xvar_sets)
for ix, xvars in enumerate(xvar_sets):
    ax = plt.subplot(num_plots,2,1+2*ix)
    plt.xlabel(xvars[0])
    plt.ylabel('accuracy')
    for im, m in enumerate(methods):
        plt.plot(np.mean(ACCU[m][xvars[0]],axis=-1), label=m)
    ax = plt.subplot(num_plots,2,1+2*ix+1)
    for im, m in enumerate(methods):
        plt.plot(np.mean(TIME[m][xvars[0]],axis=-1), label=m)
    plt.xlabel(xvars[0])
    plt.ylabel('time')
    ax.yaxis.tick_right()
    ax.yaxis.set_label_position('right')
plt.legend()
plt.subplots_adjust(hspace=0.5)
```

```python
t = time.time()
density = 0.2

densities = np.arange(0.1, 0.9, 0.1)
methods = ['S', 'GC', 'CCM']
acc = np.zeros((len(methods),len(densities)))
for mi, method in enumerate(methods):
    for di, density in enumerate(densities):
        #print(method, density)
        res = siminf(n=20, density=density, directedness=1-density, 
                     inference_method=method, num_trials=10)
        acc[mi,di] = np.mean([res[i]['accuracy'] for i in range(len(res))])
          
print(time.time()-t)
```

```python
densities = np.arange(0.1, 0.9, 0.1)
plt.plot(densities, acc[0], label='S')
plt.plot(densities, acc[1], label='GC')
plt.legend()
plt.ylabel('accuracy')
plt.xlabel('density')
```
