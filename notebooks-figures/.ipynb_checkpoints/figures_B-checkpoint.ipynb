```python
################################################################################
# We have two hypotheses; (1) Network inference becomes harder with increasing 
# mean anti-clustering coefficient (ACC) of nodes; (2) Network inference becomes 
# harder with increasing variance of feedforward multiplicity (FFM) of edges
# and nonedges. To test both hypotheses, we compute the ACC and FFM for the ER, 
# BA, RR, and SW random-graph model.

# decide whether to include feedforward multiplicities in the computations
include_ffms = False 

t = time.time()

models = ['ER','BA','RR','SW']
dvalues = np.arange(0, 1.0, 0.05)
nvalues = np.arange(50,205,5)
num_samples = 100
fnames = ['../data/B5_feedforward_multiplicities.d',
          '../data/B5_rgm_graphs.d',
          '../data/B5_anticlustering_coefficients.d', 
          '../data/B5_accuracies.d']

if load:
    if include_ffms:
        # load feedforward multiplicities from previous run
        mults = dill.load(open(fnames[0], 'rb'))
    # load graphs from previous run
    graphs = dill.load(open(fnames[1], 'rb'))
    # load mean anticlustering coefficients  from previous run
    ccs = dill.load(open(fnames[2], 'rb')) 
    # load accuracies from previous run
    accuracies = dill.load(open(fnames[3], 'rb'))
    
else:
    
    # initialize lists and arrays
    if include_ffms:
        mults = [[[0,0] for i in range(len(nvalues))] for j in range(len(models))]
    graphs = [[0 for i in range(len(nvalues))] for j in range(len(models))]
    ccs = np.zeros((len(models), len(nvalues), num_samples))
    accuracies = np.zeros((len(models), len(nvalues), num_samples))

    for ni, n in enumerate(nvalues):    
        
        # take the time to get some indication how long this might take
        t0 = time.time() 
        
        for mi, model in enumerate(models):
            
            if include_ffms:
                # initialize local list of feedforward multiplicities
                ffms = [[0 for i in range(num_samples)], 
                        [0 for i in range(num_samples)]]
                
            # initialize local list of graphs
            gs = [0 for i in range(num_samples)]
            
            for i in range(num_samples):
                
                # generate graph and get adjacency matrix
                A = make_adjacency(n, 0.1, model=model)
                
                if include_ffms:
                    # get feedforward multiplicities for edges and non-edges
                    ffm_e, ffm_ne = feedforward_multiplicities(A, add_wedges=False)
                    ffms[0][i] = ffm_e
                    ffms[1][i] = ffm_ne
                
                # get graph from adjacency matrix
                gs[i] = nx.from_numpy_array(A, create_using=nx.DiGraph)
                
                # create undirected copy of the graph
                g_undirected = nx.Graph(gs[i])
                
                # compute its degree sequence
                degrees = g_undirected.degree
                
                # compute its clustering coefficients
                clustering = nx.clustering(g_undirected)
                
                # compute its mean anti-clustering coefficient
                ccs[mi,ni,i] = np.nanmean([(1-clustering[i])*degrees[i] 
                                           if degrees[i]>0 else np.nan
                                           for i in clustering.keys()])
                accuracies[mi,ni,i] = siminf(A)['accuracy']

            # store results in arrays and lists
            #ccs[mi,ni] = np.nanmean(ccs_arr)                    
            graphs[mi][ni] = [g for g in gs]
            if include_ffms:
                mults[mi][ni][0] = [m for m in ffms[0]]
                mults[mi][ni][1] = [m for m in ffms[1]]
            
        print(n, time.time()-t0)
 
    # save data
    if include_ffms:
        dill.dump(mults, open(fnames[0], 'wb'))
    dill.dump(graphs, open(fnames[1], 'wb'))
    dill.dump(ccs, open(fnames[2], 'wb'))
    dill.dump(accuracies, open(fnames[3], 'wb'))
    
print('time', time.time()-t)
```

```python
################################################################################
# Combine the feedforward multiplicities for edges and non-edges in a single
# measure.

if include_ffms:
    # initialize array
    mults_nz2 = np.zeros((len(models),len(nvalues),4))

    for mi, model in enumerate(models):
        for ni, n in enumerate(nvalues):    
            t0 = time.time()

            res = np.zeros(100)
            for j in range(100):
                ffm_e = sorted(mults[mi][ni][0][j])
                ffm_ne = mults[mi][ni][1][j]

                lffm_ne = sorted(ffm_ne)[-len(ffm_e):]
                offset = np.sum(ffm_e > lffm_ne[-1])

                ffm_e = ffm_e[:len(ffm_e)-offset]
                lffm_ne = lffm_ne[offset:]

                res[j] = (np.mean(ffm_e)-np.mean(lffm_ne))#/(np.mean(ffm_e))

            mults_nz2[mi][ni] = np.mean(res) #arr, axis=1)
```

```python
################################################################################
# Finally, we look at correlation coefficients for anti-clustering coefficients 
# and accuracy across networks.

from scipy.stats import pearsonr, spearmanr

print('Overall Spearman R', 
      spearmanr(np.ravel(accuracies[:,:,:]), np.ravel(ccs[:,:,:])))
p1a = np.zeros(4)
p1b = np.zeros(31)
p2 = np.zeros((4,31))

for i in range(4):
    p1a[i] = spearmanr(np.ravel(accuracies[i,:]), np.ravel(ccs[i,:]))[0]
    for j in range(31):
        if i==0:
            p1b[j] = spearmanr(np.ravel(accuracies[:,j]), np.ravel(ccs[:,j]))[0]
        p2[i,j] = spearmanr(accuracies[i,j], ccs[i,j])[0]
        
print('Mean R', np.nanmean(p2))
print('Mean R', np.nanmean(p1a))
print('Mean R', np.nanmean(p1b))
```
